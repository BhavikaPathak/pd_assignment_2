```python
def reindex_with_increment(df):

    new_index = pd.Series(range(1, 2 * len(df), 2))

    new_df = df.copy()
    new_df.reset_index(drop=True, inplace=True)
    new_df.index = new_index

    return new_df
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': ['apple', 'banana', 'cherry', 'date', 'elderberry']
}
df = pd.DataFrame(data)
new_df = reindex_with_increment(df)
print("Data frame without re-index function")
print(df)
print("Data frame with re-index function")
print(new_df)
```

```python
text_data = [
    "MY NAME IS ABC",
    "I WANT TO BEACOME A DATA SCIENTIST",
    "My expertise lies in data modelling, data mining, and statistical analysis",
    "I'm proficient in Python, SQL"
]
df = pd.DataFrame({'Text': text_data})
print("Dataframe without word_count column")
print(df)

def count_words(text_data):

    words = str(text_data).split()
 
    return len(words)

def add_word_count_column(df):
    df['Word_Count'] = df['Text'].apply(count_words)
    return df
df = add_word_count_column(df)
print("Dataframe with word_count column")
print(df)
```

```python
data = {'A':[3,8,6,2,9],'B':[5,2,9,3,1],'C':[1,7,4,5,2]}
df = pd.DataFrame(data)
print(df)
def filter_dataframe(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows
new_df = filter_dataframe(df)

print("A new DataFrame that contains only the selected rows are --------> Column 'A' is greater than 5 and the value in column 'B' is less than 10")
print(new_df)
```

#                                           ANSWER 9 

```python
index = [1,2,3,4,5,6,7,8,9]
data = {'Values': [11,22,33,44,55,66,77,88,99]}
df = pd.DataFrame(data,index=index)
print(df)

def calculate_statistics(df):
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_deviation = df['Values'].std()
    return mean_value, median_value, std_deviation
mean_val, median_val, std_dev = calculate_statistics(df)

print("Mean: ", mean_val)
print("Median: ", median_val)
print("Standard Deviation: ", std_dev)
```

# ANSWER 10

```python
data = {'Sales':[1000, 1500, 800, 1200, 2000],'Date':['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04', '2023-07-05']}
df = pd.DataFrame(data)
print(df)

def calculate_moving_average(df):
    # Calculate the moving average using a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df
df = calculate_moving_average(df)
print(df)
```

#                                            ANSWER 11 

```python
data = {'Date': ['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05','2023-01-06','2023-01-07']}
df = pd.DataFrame(data)
print(df)

def add_weekday_column(df):
    
    df['Date'] = pd.to_datetime(df['Date'])
    
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df
df = add_weekday_column(df)
print(df)
```

```python
timestamp_data = ['2023-01-10 12:30:00', '2023-01-15 15:45:00', '2023-02-05 10:00:00', '2023-01-20 08:00:00','2023-08-10 12:30:00', '2023-05-15 15:45:00', '2023-03-05 10:00:00', '2023-01-22 08:00:00']
date_values = pd.to_datetime(timestamp_data)
df = pd.DataFrame({'Date': date_values})
print(df)

def select_rows_between_dates(df, start_date, end_date):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows
start_date = '2023-01-01'
end_date = '2023-01-31'
selected_df = select_rows_between_dates(df, start_date, end_date)
print(selected_df)
```
